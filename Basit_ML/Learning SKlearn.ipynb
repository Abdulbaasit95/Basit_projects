```python
import numpy as np
import pandas as pds
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

fname = "train.csv"
```

```python
from sklearn.datasets import load_iris

dat = load_iris()

features = pds.DataFrame(dat.data)
features.columns = dat.feature_names

target = pds.DataFrame(dat.target)
target.columns = pds.Series("Species")

data = pds.concat([features, target], axis= 1)
sns.pairplot(data, hue = "Species", height=2)
```
